```python

# Função para classificar a faixa de tempo
def classificar_faixa(tempo):
    if tempo <= 2:
        return 'de 0 a 2 dias'
    elif 3 <= tempo < 5:
        return 'de 3 a 5 dias'
    elif 5 <= tempo < 10:
        return 'de 5 a 10 dias'
    elif 10 <= tempo < 15:
        return 'de 10 a 15 dias'
    elif 15 <= tempo <= 20:
        return 'de 15 a 20 dias'
    elif 20 < tempo:
        return 'mais que 20'


# Aplicar a função à coluna "Tempo sem atribuição" e criar a coluna "Faixa de tempo"
base['Faixa de tempo'] = base['Tempo sem atribuição'].apply(lambda x: classificar_faixa(x))

# Função para classificar a faixa de tempo
def classificar_duracao(duracao):
    if duracao <= 25:
        return 'de 0 a 25 dias'
    elif 26 <= duracao < 50:
        return 'de 25 a 50 dias'
    elif 50 <= duracao < 75:
        return 'de 50 a 75 dias'
    elif 75 <= duracao < 100:
        return 'de 75 a 100 dias'
    elif 100 <= duracao <= 150:
        return 'de 100 a 150 dias'
    elif 150 <= duracao <= 200:
        return 'de 100 a 150 dias'
    elif 200 < duracao:
        return 'mais que 200'


# Aplicar a função à coluna "Tempo sem atribuição" e criar a coluna "Faixa de tempo"
base['Faixa de tempo de duração'] = base['Tempo de duração da SUP'].apply(lambda y: classificar_duracao(y))


cd99 = ((base['Local'] == 'EQTL Goiás'))

base.loc[cd99, 'Localização'] = 'Goiás'
base.loc[~cd99, 'Localização'] = 'Corporativo'

# Exporta a Base para um arquivo excel
base.to_excel('Base.xlsx')

```
